El conjunto de datos CIFAR10

El conjunto de datos CIFAR-10 consta de 60000 $32 \times 32$ imágenes en color en 10 clases, con 6000 imágenes por clase. Hay 50000 imágenes de entrenamiento y 10000 imágenes de prueba.

El conjunto de datos se divide en cinco lotes de entrenamiento y un lote de prueba, cada uno con 10000 imágenes. El lote de prueba contiene exactamente 1000 imágenes seleccionadas al azar de cada clase. Los lotes de entrenamiento contienen las imágenes restantes en orden aleatorio, pero algunos lotes de entrenamiento pueden contener más imágenes de una clase que de otra. Entre ellos, los lotes de entrenamiento contienen exactamente 5000images from each class.</li>

Aquí están las clases en el conjunto de datos, así como 10 imágenes aleatorias de cada una:
